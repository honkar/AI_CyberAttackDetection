### **IDENTIFY COLUMNS FOR ENCODING**


```
all_cols=df_train.columns
cat_cols=["proto","state","service","is_sm_ips_ports","is_ftp_login","attack_cat","label"]
bin_cols=["is_sm_ips_ports","is_ftp_login","label"]
num_cols=[]
cat_enc_cols=[]
```

#### **Columns for Numeric Encoding**

```
for c in all_cols:
  if c not in cat_cols:
    num_cols.append(c)

print(num_cols)
```

#### **Columns for Categorical Encoding**

```
for d in cat_cols:
  if d not in bin_cols:
    cat_enc_cols.append(d)

print(cat_enc_cols)
```

#### **Column details**

```
totalSize=0
for i in cat_enc_cols:
  size=len(df_train[i].unique())
  print()
  print(i, " :")
  print("size: ",size )
  print(df_train[i].unique())
  totalSize+=size

print("Total Size: " , totalSize)
print("Total Categorical columns: ", totalSize + len(bin_cols))
print("Total Numeric columns: ", len(num_cols))
print("Total columns: " , totalSize + len(bin_cols) + len(num_cols))
```

### **Identify common columns and drop other columns**

```
train_allCols=df_train.columns
test_allCols=df_test.columns
common_cols=set(train_allCols).intersection(set(test_allCols))
drop_trainCols=[]
drop_testCols=[]

for c in train_allCols:
  if c not in common_cols:
    drop_trainCols.append(c)

for c in test_allCols:
  if c not in common_cols:
    drop_testCols.append(c)

print("Total columns in train set: ", len(train_allCols))
print("Total columns in test set: ", len(test_allCols))
print("Additional columns in train set: ", len(drop_trainCols))
print("Additional columns in test set: ", len(drop_testCols))
print("Common columns: ", len(common_cols))
```

Drop additional columns

```
for c in drop_trainCols:
  df_train=df_train.drop(c, axis=1)

for d in drop_testCols:
  df_test=df_test.drop(d, axis=1)
```
